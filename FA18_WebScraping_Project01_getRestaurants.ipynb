<h3><center>PERFORMING ANALYSIS OF THE EXTRACTED DATA </h3></center>

#### FILE USED : RestaurantData.csv

````python
import pandas as pd

## KINDLY UPDATE THE NAME AND PATH OF THE OUTPUT .csv IN THIS CELL IF YOU WISH TO RE-EXECUTE THIS CODE

file = 'RestaurantData.csv'

df = pd.read_csv(file, encoding = "ISO-8859-1", names=['Name','Link','Rating','NumOfReviews','PriceCategory','Categories','Phone','Address','Locality'])
df.head()

#df.info()
````

````python
#### CATEGORY WISE TOTAL REVIEWS
# df[df['Categories'].str.contains("American \(New\)")]['NumOfReviews'].replace(' ','0').astype(int).sum()
cat_list = []
for i in df['Categories']:
    i=i.replace("[",'').replace("]",'').replace('\'','')
    i=i.split(', ')
    for j in i:
        cat_list.append(j)
cat_set = set(cat_list)
dict_reviewCount={}
for i in cat_set:
    cat = i.replace('(','\(').replace(')','\)')
    reviews = df[df['Categories'].str.contains(cat)]['NumOfReviews'].replace(' ','0').astype(int).sum()
    dict_reviewCount[i] = reviews
#print(dict_reviewCount) 

#### MOST REVIEWED CATEGORY
df_reviewCount = pd.DataFrame.from_dict(dict_reviewCount,orient='index',columns=['reviews'])
print("MOST REVIEWED RESTAURANT CATEGORY :\n", df_reviewCount[df_reviewCount['reviews']==df_reviewCount['reviews'].max()])
print()
#### LEAST REVIEWED CATEGORY
print("LEAST REVIEWED RESTAURANT CATEGORY :\n", df_reviewCount[df_reviewCount['reviews']==df_reviewCount['reviews'].min()])
````

````python
import matplotlib.pyplot as plt
%matplotlib inline

print("PLOT OF 'Resturant categories Vs Number of reviews' FOR 50 FIRST CATEGORIES : " )
df_reviewCount.sort_values('reviews', ascending=False,inplace=True)

plt.style.use('ggplot')
plt.figure(figsize=[25,10])
plt.bar(df_reviewCount.index[:50],df_reviewCount['reviews'][:50])
plt.xlabel('Categories')
plt.ylabel('Number of reviews')
plt.xticks(rotation=90)
plt.title('Resturant categories Vs Number of reviews')
plt.show()
````

````python
print("LIST OF RESTAURANTS WITH HIGHEST RATING : ")
df[['Name','Rating','NumOfReviews']][df['Rating']==df['Rating'].max()]
````

````python
## RESTO WITH MAX REVIEWS
# df.sort_values('NumOfReviews',ascending=False)
print("LIST OF RESTAURANTS WITH MAXIMUM REVIEWS : ")
df_new = df
df_new['NumOfReviews'] = df_new['NumOfReviews'].replace(' ','0').replace('','0').astype(int)
df_new['Categories'] = df_new['Categories'].str.replace('\[|\'|\]','')
df_new.sort_values('NumOfReviews',ascending=False)[:5][['Name','NumOfReviews','Rating','Categories']]
````

````python
import matplotlib.pyplot as plt
%matplotlib inline

df_priceCat = df
df_priceCat['PriceCategory']=df_priceCat['PriceCategory'].replace(' ','Unknown')

print("ANALYSIS OF NUMBER OF REVIEWS FOR PRICE CATEGORIES : ")
print(df_priceCat.groupby('PriceCategory')['NumOfReviews'].sum())
print("PRICE CATEGORIES : " , list(df_priceCat.groupby('PriceCategory')['NumOfReviews'].sum().index))
print("TOTAL REVIEWS FOR PRICE CATEGORIES : " , list(df_priceCat.groupby('PriceCategory')['NumOfReviews'].sum()[:]))
print()
print("PLOT OF 'Price categories Vs Total Number of reviews' : " )

#plt.figure(figsize=[25,10])
plt.bar(list(df_priceCat.groupby('PriceCategory')['NumOfReviews'].sum().index),list(df_priceCat.groupby('PriceCategory')['NumOfReviews'].sum()[:]))
plt.xlabel('Price Categories')
plt.ylabel('Total Number of reviews')
plt.xticks(rotation=0)
plt.title('Price categories Vs Total Number of reviews')
plt.show()
````
